```python
# Python program to read image using OpenCV

# importing OpenCV(cv2) module
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
from numpy import array
from matplotlib import pyplot as plt
img = cv2.imread('/content/IMAGE.jpg')
cv2_imshow(img)
```

```python
# Check the image matrix data type (could know the bit depth of the image)
print(img.dtype)
# Check the height of image
print(img.shape[0])
# Check the width of image
print(img.shape[1])
# Check the number of channels of the image
print(img.shape[2])

# details about the image printed below
print(type(img))
print(img.size)

#Convert Image To Matrix
ar = array(img)
ar
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
ar = array(gray_image)
ar
```

```python
#histogram of color image
plt.hist(img.ravel(),256,[0,256])
plt.show()

# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])

# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
#histogram of grayscale image
plt.hist(gray_image.ravel(),256,[0,256])
plt.show()

# find frequency of pixels in range 0-255
histr = cv2.calcHist([gray_image],[0],None,[256],[0,256])

# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

A grayscale image can be represented as as two dimensional array, whose first axis corresponds to the x coordinate of the image and the second axis corresponds to the y coordinate. The array contains at each coordinate pair (x,y) a value, which is typically a float between 0.0 and 1.0, or an integer between 0 and 255.
