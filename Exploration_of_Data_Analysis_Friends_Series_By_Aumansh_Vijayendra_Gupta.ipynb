```python
# Create a new figure with specified size (width=10, height=5)
plt.figure(figsize = (10, 5))

# Create a count plot using Seaborn
# x="Season": Plot the "Season" column on the x-axis
# data=friends: Use the "friends" DataFrame for data
# palette='inferno': Set the color palette to 'inferno'
sns.countplot(x = "Season", data = friends, palette = 'inferno')

# Set the title of the plot
plt.title("Number of Episodes per Season")

# Set the label for the x-axis
plt.xlabel("Season")

# Set the label for the y-axis
plt.ylabel("Count")

# Display the plot
plt.show()
```

```python
Season_Duration = friends.groupby('Season')['Duration'].sum().reset_index()

Season_Duration
```

```python
plt.figure(figsize = (10,5))

sns.barplot(x = Season_Duration['Season'], y = Season_Duration['Duration'], palette = 'rocket')

plt.title('Total Duration of each Season', fontsize = 15)

plt.xlabel('Duration')

plt.ylabel('Season')

plt.xticks(fontsize = 12)

plt.yticks(fontsize = 12)

plt.ylim(300, 600);
```

```python
# Find the row with the highest duration
highest_duration_row = friends.loc[friends['Duration'].idxmax()]
Highest_Duration = highest_duration_row['Duration']
Highest_Duration_Episode = highest_duration_row['Episode_Title']
Highest_Duration_Season = highest_duration_row['Season']

# Find the row with the least duration
least_duration_row = friends.loc[friends['Duration'].idxmin()]
Least_Duration = least_duration_row['Duration']
Least_Duration_Episode = least_duration_row['Episode_Title']
Least_Duration_Season = least_duration_row['Season']

# Print the results with proper formatting
print(f"Highest Duration: {Highest_Duration} minutes | Episode: {Highest_Duration_Episode} | Season: {Highest_Duration_Season}")
print(f"Least Duration: {Least_Duration} minutes | Episode: {Least_Duration_Episode} | Season: {Least_Duration_Season}")
```

```python
Season_Stars_Total = friends.groupby("Season")['Stars'].sum().reset_index()

Season_Stars_Total.columns = ['Season', 'Total Stars']

Season_Stars_Total
```

```python
plt.figure(figsize = (10, 5))

sns.lineplot(x = 'Season', y = 'Total Stars', data = Season_Stars_Total, marker ='v')

plt.title('Total Star Rating For Each Season', fontsize=15)

plt.xlabel('Total Stars')

plt.ylabel('Season')

plt.xticks(fontsize = 12)

plt.yticks(fontsize = 12);
```

```python
season_stars_avg = friends.groupby('Season')['Stars'].mean().reset_index()

season_stars_avg.columns = ['Season','Average Stars']

season_stars_avg
```

```python
plt.figure(figsize = (10,5))

sns.lineplot(x = 'Season', y = 'Average Stars', data = season_stars_avg, marker="o")

plt.title('Average Star Rating For Each Season', fontsize=15)

plt.xlabel('Average Stars')

plt.ylabel('Season')

plt.xticks(fontsize=12)

plt.yticks(fontsize=12);
```

```python
highest_stared_episode = friends[['Episode_Title', 'Stars', 'Season']].sort_values('Stars', ascending=False).head(10).reset_index(drop = True)
highest_stared_episode
```

```python
plt.figure(figsize = (10,6))

sns.barplot(x = 'Stars', y = 'Episode_Title', data = highest_stared_episode, palette = 'rocket', orient = 'h')

plt.title('Top 10 High-Rated Episodes', fontsize = 15)

plt.xlabel('Stars')

plt.ylabel('Episodes')

plt.xticks(fontsize = 12)

plt.yticks(fontsize = 12)

plt.xlim(9, 10);
```

```python
recurring_director_count = friends.groupby("Director")['Episode_Title'].count().sort_values(ascending = False)
recurring_director_count
```

```python
print(f"There are {recurring_director_count.count()} directors.")
```

```python
recurring_director_10ep = recurring_director_count.reset_index()

recurring_director_10ep = recurring_director_10ep[recurring_director_10ep["Episode_Title"]>=10]
recurring_director_10ep
```

```python
plt.figure(figsize = (20, 8))

sns.barplot(x = recurring_director_10ep['Director'], y = recurring_director_10ep['Episode_Title'], palette = 'Oranges')

plt.title('Top 10 Recurring Directors', fontsize = 15)

plt.xlabel('Directors', fontsize = 15)

plt.ylabel('No of Episodes Directed')

plt.xticks(fontsize = 13)

plt.yticks(fontsize = 13);
```

```python
recurring_director_avg_rating = friends.groupby("Director").agg({"Episode_Title": "count", "Stars":"mean"}).sort_values(by = "Stars", ascending=False)
recurring_director_avg_rating

recurring_director_10_avg_rating = recurring_director_avg_rating[recurring_director_avg_rating["Episode_Title"]>=10]
recurring_director_10_avg_rating
```

```python
plt.figure(figsize = (15, 10))

sns.lineplot(x = 'Director', y = 'Stars', data = recurring_director_10_avg_rating, marker = 'H')

plt.title('Recurring Director\'s Average Star Rating')

plt.xlabel('Directors')

plt.ylabel('Average Stars Across Episodes');
```

```python
director_rating = friends.groupby("Director")['Stars'].agg(['count','mean']).sort_values(by = "mean", ascending = False).reset_index()

dr_85 = director_rating[director_rating['mean'] >= 8.5]
dr_85
```

```python
plt.figure(figsize = (20, 10))

sns.barplot(x = 'Director', y = 'mean', data = dr_85, palette = 'rocket')

plt.title('DIRECTORS WITH AT LEAST AVERAGE STAR RATING OF THE  8.5 STARS', fontsize = 15)

plt.xlabel('Directors', fontsize = 15)

plt.ylabel('Average Stars Across Episodes', fontsize = 15)

plt.ylim(8.3, 9.3);
```

```python
Season_Votes = friends.groupby("Season")['Votes'].sum().reset_index()
Season_Votes
```

```python
plt.figure(figsize = (10, 8))

sns.barplot(x = 'Season', y = 'Votes', data = Season_Votes, palette = 'magma')

plt.title('Total Votes For Each Season')

plt.xlabel('Seasons')

plt.ylabel('Total Votes')

plt.ylim(60000, 110000);
```

```python
recurring_director_votes = friends.groupby("Director").agg({"Episode_Title":"count", "Votes": "sum"}).sort_values(by = "Votes", ascending = False).reset_index()

recurring_director_votes_10 = recurring_director_votes[recurring_director_votes["Episode_Title"] >= 10]
recurring_director_votes_10
```

```python
plt.figure(figsize = (25, 12))

sns.barplot(y = 'Votes', x = 'Director', data = recurring_director_votes_10, palette='inferno')

plt.ylabel('Votes', fontsize = 20)
plt.xlabel('Director', fontsize = 20)
plt.title('Top High-voted Directors (Directing more than 10 Episodes)', fontsize = 20)
plt.ylim(20000, 200000)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18);
```
